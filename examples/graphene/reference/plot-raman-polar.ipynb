```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import numpy as np

def IRaman(theta,R):
    Pi = np.array([[np.cos(theta)], [np.sin(theta)], [0]])
    Ps = np.array([np.cos(theta), np.sin(theta), 0])
    I = Ps.dot(R).dot(Pi)
    return abs(I)**2

#Raman tensor for test
#np.zeros((3,3), dtype=complex)
R = [[0.1+1j,0+0j,0+0j],[0+0j,0.1-1j,0+0j],[0+0j,0+0j,0+0j]]

#Complex Raman tensor from raman.out
R1 = [[0.11793505E-001+0.68925106E-002j,0.10368675E+000+0.99460610E+000j,0.27814843E-004-0.35557944E-005j],
     [0.10368476E+000+0.99461021E+000j,0.14858385E-002-0.75841993E-002j,-0.11568847E-005-0.14581073E-004j],
     [0.26533987E-004+0.74147289E-005j,-0.56400067E-005-0.45370302E-007j,0.48303692E-003-0.38809866E-005j]]

R2 = [[0.10460614E+000+0.99386909E+000j,0.35315700E-002-0.76057291E-002j,-0.25792587E-005-0.10850621E-005j],
     [0.35383251E-002-0.76049908E-002j,-0.11627372E+000-0.99321721E+000j,0.13699245E-004-0.91544268E-005j],
     [-0.55064671E-005-0.87526182E-005j,0.28277079E-004-0.18703235E-005j,-0.28230643E-003+0.22623184E-005j]]

theta = np.arange(0,(2*np.pi),0.01)

# Creating a new figure and setting up the resolution
fig = plt.figure()

# Change the coordinate system from scaler to polar
ax = fig.add_subplot(projection='polar')
ax.set_yticklabels([])
ax.tick_params(axis='x', which='major', pad=18)

# Plotting the polar coordinates
for i in theta:
    plt.polar(i,IRaman(i,R1),'C1.')

# Save a figure to the pdf file
plt.savefig('polar-raman1.pdf')
# display the Polar plot
plt.show()
```
