```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import numpy as np

# Open and read the raman_spectra file
omega1, raman1 = np.genfromtxt('./36x36x1/raman_spectra1.dat', usecols=(0,4), skip_footer=1, unpack=True)
omega2, raman2 = np.genfromtxt('./36x36x1/raman_spectra1.dat', usecols=(0,8), skip_footer=1, unpack=True)

omega3, raman3 = np.genfromtxt('./48x48x1/raman_spectra1.dat', usecols=(0,4), skip_footer=1, unpack=True)
omega4, raman4 = np.genfromtxt('./48x48x1/raman_spectra1.dat', usecols=(0,8), skip_footer=1, unpack=True)

omega5, raman5 = np.genfromtxt('./60x60x1/raman_spectra1.dat', usecols=(0,4), skip_footer=1, unpack=True)
omega6, raman6 = np.genfromtxt('./60x60x1/raman_spectra1.dat', usecols=(0,8), skip_footer=1, unpack=True)

omega7, raman7 = np.genfromtxt('./72x72x1/raman_spectra1.dat', usecols=(0,4), skip_footer=1, unpack=True)
omega8, raman8 = np.genfromtxt('./72x72x1/raman_spectra1.dat', usecols=(0,8), skip_footer=1, unpack=True)

# Create figure object
plt.figure(figsize=(6,5))

plt.plot(omega2, raman2/raman2.max(), c='C0', ls='-', label=r'$36\times 36 \times 1$')
plt.plot(omega1, raman1/raman2.max(), c='C0', ls='-')

plt.plot(omega4, raman4/raman2.max(), c='C1', ls='--', label=r'$48\times 48 \times 1$')
plt.plot(omega3, raman3/raman2.max(), c='C1', ls='--')

plt.plot(omega6, raman6/raman2.max(), c='C2', ls=':', label=r'$60\times 60 \times 1$')
plt.plot(omega5, raman5/raman2.max(), c='C2', ls=':')

plt.plot(omega8, raman8/raman2.max(), c='C4', ls='-.', label=r'$72\times 72 \times 1$')
plt.plot(omega7, raman7/raman2.max(), c='C4', ls='-.')

plt.text(1530, 0.3, r'$\sigma{+}\sigma{-}$', fontsize = 22)
plt.text(1530, -0.1, r'$\sigma{+}\sigma{+}$', fontsize = 22)

# Add the x and y-axis labels
plt.xlabel(r'Raman shift (cm$^{-1}$)')
plt.ylabel(r'Raman intensity (arb. units)')
plt.yticks([])
# Add the the legend
plt.legend(loc='upper right',ncol=1)
# Set the axis limits
plt.xlim(1500, 1580)
plt.ylim(-0.2,1.1)
# Save a figure to the pdf file
plt.savefig('raman-spectra.pdf')
# Show plot
plt.show()
```
