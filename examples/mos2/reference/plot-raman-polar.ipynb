```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import numpy as np

def IRaman(theta,R):
    Pi = np.array([[np.cos(theta)], [np.sin(theta)], [0]])
    Ps = np.array([np.cos(theta), np.sin(theta), 0])
    I = Ps.dot(R).dot(Pi)
    return abs(I)**2

#Raman tensor for test
#np.zeros((3,3), dtype=complex)
R = [[0.1+1j,0+0j,0+0j],[0+0j,0.1-1j,0+0j],[0+0j,0+0j,0+0j]]

#Raman tensor (image part) from raman.out for 2.54 eV
R1 = [[0.22986840E-001-0.16744782E-001j, -0.56967379E+000-0.81263534E+000j, 0.23499029E-001-0.11026787E-001j],
     [-0.57466724E+000-0.81838717E+000j, 0.27483757E-001+0.58851200E-002j, 0.34953517E-004+0.23791352E-004j],
     [0.23583429E-001-0.11836789E-001j, -0.11879159E-003-0.10148777E-003j, 0.32201686E-003+0.17864744E-004j]]

R2 = [[0.57991407E+000+0.81467765E+000j, 0.19227110E-002+0.11600095E-001j, -0.15374454E-005+0.15338484E-003j],
     [0.84395440E-003+0.76888431E-002j, -0.55112655E+000-0.81902752E+000j, -0.23032377E-001+0.10890336E-001j],
     [-0.11072911E-003+0.11274261E-003j, -0.23187363E-001+0.11779041E-001j, 0.44684104E-003-0.26903616E-004j]]

R3 = [[0.12106493E+000-0.99264459E+000j, 0.76728848E-002+0.15638659E-001j, -0.19464393E-004-0.98219119E-004],
     [0.12011558E-001+0.10961757E-001j, 0.13337798E+000-0.97660986E+000j, 0.87420952E-004+0.91842962E-005],
     [-0.11491541E-003-0.86242954E-005j, -0.73993565E-004-0.39269713E-004j, 0.90212962E+000-0.40007123E+000]]

theta = np.arange(0,(2*np.pi),0.01)

# Creating a new figure and setting up the resolution
fig = plt.figure()

# Change the coordinate system from scaler to polar
ax = fig.add_subplot(projection='polar')
ax.set_yticklabels([])
ax.tick_params(axis='x', which='major', pad=18)

# Setting the axis limit
ax.set_ylim(0,1.2)

# Plotting the polar coordinates
for i in theta:
    plt.polar(i,IRaman(i,R3),'C4.')

# Save a figure to the pdf file
plt.savefig('polar-raman3.pdf')
# display the Polar plot
plt.show()
```
