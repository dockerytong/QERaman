```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import numpy as np

# Open and read the raman_spectra file
omega1, raman1 = np.genfromtxt('./raman_spectra1.dat', usecols=(0,4), skip_footer=1, unpack=True)
omega2, raman2 = np.genfromtxt('./raman_spectra1.dat', usecols=(0,8), skip_footer=1, unpack=True)

omega3, raman3 = np.genfromtxt('./raman_spectra2.dat', usecols=(0,4), skip_footer=1, unpack=True)
omega4, raman4 = np.genfromtxt('./raman_spectra2.dat', usecols=(0,8), skip_footer=1, unpack=True)

omega5, raman5 = np.genfromtxt('./raman_spectra3.dat', usecols=(0,4), skip_footer=1, unpack=True)
omega6, raman6 = np.genfromtxt('./raman_spectra3.dat', usecols=(0,8), skip_footer=1, unpack=True)

omega7, raman7 = np.genfromtxt('./raman_spectra4.dat', usecols=(0,4), skip_footer=1, unpack=True)
omega8, raman8 = np.genfromtxt('./raman_spectra4.dat', usecols=(0,8), skip_footer=1, unpack=True)


# Create figure object
plt.figure(figsize=(5,10))

plt.plot(omega2, raman2/raman1.max(), c='C0', ls='--', label=r'$\sigma{+}\sigma{-}$')
plt.plot(omega1, raman1/raman1.max(), c='C1', ls='-', label=r'$\sigma{+}\sigma{+}$')

plt.plot(omega4, raman4/raman1.max()+1.5, c='C0', ls='--')
plt.plot(omega3, raman3/raman1.max()+1.5, c='C1', ls='-')

plt.plot(omega6, raman6/raman1.max()+3, c='C0', ls='--')
plt.plot(omega5, raman5/raman1.max()+3, c='C1', ls='-')

plt.plot(omega8, raman8/raman1.max()+4.5, c='C0', ls='--')
plt.plot(omega7, raman7/raman1.max()+4.5, c='C1', ls='-')


plt.text(361, 0.15, r'1.95 eV', fontsize = 20)
plt.text(361, 1.65, r'2.33 eV', fontsize = 20)
plt.text(361, 3.15, r'2.54 eV', fontsize = 20)
plt.text(361, 4.65, r'2.87 eV', fontsize = 20)
plt.text(378, 6.3, r'$\sigma{+}\sigma{-}$', fontsize = 20)
plt.text(398, 5.6, r'$\sigma{+}\sigma{+}$', fontsize = 20)

# Add the x and y-axis labels
plt.xlabel(r'Raman shift (cm$^{-1}$)')
plt.ylabel(r'Raman intensity (arb. units)')
plt.yticks([])
# Add the the legend
#plt.legend(loc='upper left',ncol=1)
# Set the axis limits
plt.xlim(360, 420)
plt.ylim(-1,7)
# Save a figure to the pdf file
plt.savefig('raman-spectra.pdf')
# Show plot
plt.show()
```
